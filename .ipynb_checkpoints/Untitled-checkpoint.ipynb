```python
from sklearn.datasets import load_iris
import numpy as np
iris = load_iris()
```

```python
X = iris.data
X[:10, :]
```

```python
scales = np.amax(X, axis = 0)
```

```python
# Basic Fuzzy c-means clustering 
c = 3
epsilon  = .05
m = 2
V = np.empty((c,4))
for i in np.arange(0,c):
    c_ = np.random.random((c,1))* scales[i]
    V[:, i:i+1] = c_
    
fuzz = np.random.random(X.shape)    
```
