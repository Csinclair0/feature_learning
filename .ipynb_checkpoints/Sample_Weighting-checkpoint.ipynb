# Feature Weighted Fuzzy C-means Classifier Visuals

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
n_points_per_cluster = 250

C1 = [0, 1] + .4 * np.random.randn(n_points_per_cluster, 2)
C2 = [1, 2] + .25 * np.random.randn(n_points_per_cluster, 2)
C3 = [1, 3] + .5 * np.random.randn(n_points_per_cluster, 2)

C1 = np.hstack((C1, 1 * np.random.randn(n_points_per_cluster, 1)))
C2 = np.hstack((C2, 1 * np.random.randn(n_points_per_cluster, 1)))
C3 = np.hstack((C3, 1 * np.random.randn(n_points_per_cluster, 1)))
from mpl_toolkits.mplot3d import Axes3D
fig3D = plt.figure(figsize=(20, 10))
ax = fig3D.add_subplot(121, projection='3d')
ax.scatter(C1[:, 0], C1[:, 1],C1[:, 2], c= 'red',  alpha = .5)
ax.scatter(C2[:, 0], C2[:, 1],C2[:, 2], c= 'green',  alpha = .5)
ax.scatter(C3[:, 0], C3[:, 1],C3[:, 2], c= 'blue',  alpha = .5)
ax.set_title("With Noise Feature")
ax2 = fig3D.add_subplot(122, projection='3d')
zeros = np.zeros((n_points_per_cluster, 1))
ax2.scatter(C1[:, 0], C1[:, 1],zeros, c= 'red',  alpha = .5)
ax2.scatter(C2[:, 0], C2[:, 1],zeros, c= 'green',  alpha = .5)
ax2.scatter(C3[:, 0], C3[:, 1],zeros, c= 'blue',  alpha = .5)
ax2.set_title("Without Noise Feature")
plt.savefig("NoiseReduction.png")
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig3D = plt.figure(figsize=(15, 15))
ax = fig3D.add_subplot(111, projection='3d')
zeros = np.zeros((n_points_per_cluster, 1))
ax.scatter(C1[:, 0], C1[:, 1],zeros, c= 'red',  alpha = .5)
ax.scatter(C2[:, 0], C2[:, 1],zeros, c= 'green',  alpha = .5)
ax.scatter(C3[:, 0], C3[:, 1],zeros, c= 'blue',  alpha = .5)
```
